# Come√ßando aqui Regress√£o

```python
df.info()
```

```python
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df.columns
```

Verificando a correla√ß√£o com as demais colunas

```python
df.corr()['Income'].sort_values()
```

Aqui inserimos esse comando apenas para facilitar e copiar o nome das colunas que tem mais correla√ß√£o pois foi colocado em ordem.

```python
df.corr()['Income'].sort_values().index
```

' NumWebVisitsMonth' , 'NumStorePurchases', 'MntWines', 'MntMeatProducts', 'NumCatalogPurchases'

Aqui separamos os n√£o nulos para a regress√£o.

X recebe todas as colunas sem Income e sem dados correspondentes aos nulos da coluna Income.

Y recebe apenas Income n√£o nulo.

```python
lr = LinearRegression()

df_reg = df[['NumWebVisitsMonth', 'NumStorePurchases', 'MntWines', 'MntMeatProducts', 'NumCatalogPurchases', 'Income']]
df_reg= df_reg[df_reg["Income"].notnull()]

X = df_reg.drop(columns="Income")
y = df_reg["Income"]

lr.fit(X, y)
```

Aplicando o split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

Treinando o modelo.

```python
lr.fit(X_train,y_train)
```

Aplicando o predicti para testar o modelo e verificar os valores de erros.

```python
yHat = lr.predict(X_test)
```

```python
yHat
```

```python
from sklearn.metrics import r2_score, mean_absolute_error , mean_squared_error
```

Valor de precis√£o do teste

```python
r2 = r2_score(y_test, yHat)
print('R2 Score: ', r2)
```

Erro quadr√°tico

```python
mse = mean_squared_error(y_test, yHat)
print('MSE: ', mse)
```

Erro absoluto

```python
mae = mean_absolute_error(y_test, yHat)
print('MAE: ', mae)
```

```python
print("RMSE:", mean_squared_error(y_true=y_test, y_pred=yHat)**0.5)
```

Aplicando o modelo treinado em nossa coluna Income onde havia nulos para preencher esses valores com o que aprendeu.

```python
df_reg2 = df[['NumWebVisitsMonth', 'NumStorePurchases', 'MntWines', 'MntMeatProducts', 'NumCatalogPurchases', 'Income']]
df_reg2 = df_reg2[df_reg2["Income"].isnull()]

X2 = df_reg2.drop(columns="Income")
pred = lr.predict(X2)

df.loc[df['Income'].isnull(),'Income'] = pred






```
